```python
Y = df_copy.iloc[:,0].values
X = df_copy.iloc[:,1:].values
Y = Y.astype(int)
```

```python
Y.dtype
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y, test_size = 0.25 , random_state = 3)
```

```python
x_train.shape,x_test.shape , y_train.shape , y_test.shape
```

- Logistic regression and SVM , maybe decision tree also.
- For that , I may need to use Pipeline for each model and compare their accuracy.
- Will also use gridsearch for parameter tuning

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report,accuracy_score
```

```python
y_train
```

# Let's start with SVM

```python
svm_pipe = Pipeline([
    ('std_scaler' , StandardScaler()),
    ('svm',SVC())
])

param_grid_svm = {
    'svm__C':[0.01,0.1 , 1 ,10],
    'svm__kernel':['liner', 'rbf'],
    'svm__gamma':['scale','auto' ]
}

svm_final = GridSearchCV(svm_pipe,param_grid_svm ,cv=5,scoring = 'accuracy')
svm_final.fit(x_train , y_train)
```

```python
svm_best = svm_final.best_estimator_
svm_y_pred = svm_best.predict(x_test)
print('SVm accuracy', accuracy_score(y_test,svm_y_pred))
print(classification_report(y_test , svm_y_pred))
```

- Now lets see my best parameters for SVM

```python
print('Best Paramenters for SVM : ' ,svm_final.best_params_)
```

- Looks like C= 1 , gamma = sclae and rbf kernel shine !

# Now let's try with Logistic Regression
- I am not gonna do any para tuning since the dataset is pretty simple and doing para tuning won't be a difference at this point

```python
log_pipe = Pipeline([
    ('sceler', StandardScaler()),
    ('logreg', LogisticRegression())
])

log_pipe.fit(x_train ,y_train)
```

```python
log_y_pred = log_pipe.predict(x_test)
```

```python
print('Logistic Regression Accuracy :', accuracy_score(y_test, log_y_pred))
print('Classification report :' , classification_report(y_test, log_y_pred))
```

- Seems like log reg did better than SVM .
# Now let's see what Decision Tree can do

```python
#DT doesn't need FS ( at least I have heard so !)
dt_pipe = Pipeline([
    ('dt', DecisionTreeClassifier(random_state = 42))
])

dt_params = {
    'dt__max_depth': [2, 3, 5, 10, None],
    'dt__min_samples_leaf': [5, 10, 20, 50, 100],
    'dt__criterion': ["gini", "entropy"]
}

dt_grid = GridSearchCV(dt_pipe, dt_params, cv=5, scoring='accuracy')
dt_grid.fit(x_train, y_train)
```

```python
dt_best_model = dt_grid.best_estimator_
dt_y_pred = dt_best_model.predict(x_test)
print('Decision Tree accuracy', accuracy_score(y_test,dt_y_pred))
print(classification_report(y_test , dt_y_pred))
```

-As we can see decision tree is not the best model for this dataset.

```python
# Assuming x_test and y_test are defined and models are fitted
from sklearn.metrics import  roc_auc_score, roc_curve
from sklearn import metrics
# Get probability estimates
svm_probs = svm_best.decision_function(x_test)
log_probs = log_pipe.predict_proba(x_test)[:, 1]
dt_probs = dt_grid.best_estimator_.predict_proba(x_test)[:, 1]

# Compute ROC curves and AUCs
svm_fpr, svm_tpr, _ = roc_curve(y_test, svm_probs)
log_fpr, log_tpr, _ = roc_curve(y_test, log_probs)
dt_fpr, dt_tpr, _ = roc_curve(y_test, dt_probs)

svm_auc = metrics.auc(svm_fpr, svm_tpr)
log_auc = metrics.auc(log_fpr, log_tpr)
dt_auc = metrics.auc(dt_fpr, dt_tpr)

# Plot ROC curves
plt.figure(figsize=(10, 6))
plt.plot(svm_fpr, svm_tpr, label=f'SVM (AUC = {svm_auc:.2f})')
plt.plot(log_fpr, log_tpr, label=f'Logistic Regression (AUC = {log_auc:.2f})')
plt.plot(dt_fpr, dt_tpr, label=f'Decision Tree (AUC = {dt_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for All Models')
plt.legend()
plt.grid(True)
plt.show()
```
